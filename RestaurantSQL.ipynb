2.Is there a correlation between vendor ratings and delivery distance?


```python
%%sql
WITH mean_values AS (
    SELECT 
        AVG("vendor_rating_x") AS mean_vendor_ratings, 
        AVG("deliverydistance") AS mean_delivery_distance 
    FROM 
        "Restuarant_df"
)
SELECT 
    CORR("vendor_rating_x", "deliverydistance") AS correlation_coefficient
FROM 
    "Restuarant_df",
    mean_values;
```

```python

# SQL query
sql_query = """
WITH mean_values AS (
    SELECT 
        AVG("vendor_rating_x") AS mean_vendor_ratings, 
        AVG("deliverydistance") AS mean_delivery_distance 
    FROM 
        "Restuarant_df"
)
SELECT 
    CORR("vendor_rating_x", "deliverydistance") AS correlation_coefficient
FROM 
    "Restuarant_df",
    mean_values;
"""


# Fetch data into a pandas DataFrame
df = pd.read_sql_query(sql_query, conn)

# Close the database connection
conn.close()

# Get the correlation coefficient value from the DataFrame
correlation_coefficient = df.iloc[0]["correlation_coefficient"]

# Plot the correlation coefficient
plt.bar(["Correlation Coefficient"], [correlation_coefficient])
plt.ylabel("Correlation Coefficient")
plt.title("Correlation between Vendor Rating and Delivery Distance")
plt.show()
```

3.How does the preparation time affect the vendor's ability to deliver on time?

```python
%%sql
SELECT 
    "preparationtime",
    COUNT(*) AS total_orders,
    ROUND(AVG(EXTRACT(EPOCH FROM ("delivered_time"::timestamp - "picked_up_time"::timestamp)) / 60), 2) AS average_delivery_delay_minutes
FROM 
    "Restuarant_df"
WHERE 
    "delivered_time" IS NOT NULL
GROUP BY 
    "preparationtime"
ORDER BY 
    "preparationtime";

```

```python

# SQL query
sql_query = """
SELECT 
    "preparationtime",
    COUNT(*) AS total_orders,
    ROUND(AVG(EXTRACT(EPOCH FROM ("delivered_time"::timestamp - "picked_up_time"::timestamp)) / 60), 2) AS average_delivery_delay_minutes
FROM 
    "Restuarant_df"
WHERE 
    "delivered_time" IS NOT NULL
GROUP BY 
    "preparationtime"
ORDER BY 
    "preparationtime";
"""


# Fetch data into a pandas DataFrame
df = pd.read_sql_query(sql_query, conn)

# Close the database connection
conn.close()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(df["preparationtime"], df["total_orders"], label="Total Orders", marker="o")
plt.plot(df["preparationtime"], df["average_delivery_delay_minutes"], label="Average Delivery Delay (Minutes)", marker="o")
plt.xlabel("Preparation Time (minutes)")
plt.ylabel("Count")
plt.title("Total Orders and Average Delivery Delay by Preparation Time")
plt.legend()
plt.grid(True)
plt.show()
```

4.What is the average vendor rating?

```python
%%sql
SELECT "vendor_id",
 ROUND(AVG("vendor_rating_x")::NUMERIC, 2) AS average_vendor_rating
FROM "Restuarant_df"
GROUP BY "vendor_id"
ORDER BY average_vendor_rating DESC
LIMIT 5;
```

```python

# SQL query
sql_query = """
SELECT "vendor_id",
       ROUND(AVG("vendor_rating_x")::NUMERIC, 2) AS average_vendor_rating
FROM "Restuarant_df"
GROUP BY "vendor_id"
ORDER BY average_vendor_rating DESC
LIMIT 5;
"""


# Fetch data into a pandas DataFrame
df = pd.read_sql_query(sql_query, conn)

# Close the database connection
conn.close()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(df["average_vendor_rating"], labels=df["vendor_id"], autopct="%1.1f%%", startangle=140)
plt.title("Top 5 Vendors with Highest Average Rating")
plt.axis("equal")
plt.show()
```

7.How have vendor ratings changed over time (based on "created_at_x" or "updated_at_x")?


```python
%%sql
SELECT
    DATE_TRUNC('month', "created_at_x"::timestamp) AS time_group,
    CAST(AVG("vendor_rating_x") AS NUMERIC(10, 2)) AS average_vendor_rating
FROM
    "Restuarant_df"
GROUP BY
    DATE_TRUNC('month', "created_at_x"::timestamp)
ORDER BY
    time_group
LIMIT 10;
```

```python

# SQL query
sql_query = """
SELECT
    DATE_TRUNC('month', "created_at_x"::timestamp) AS time_group,
    CAST(AVG("vendor_rating_x") AS NUMERIC(10, 2)) AS average_vendor_rating
FROM
    "Restuarant_df"
GROUP BY
    DATE_TRUNC('month', "created_at_x"::timestamp)
ORDER BY
    time_group
LIMIT 10;
"""
# Fetch data into a pandas DataFrame
df = pd.read_sql_query(sql_query, conn)

# Close the database connection
conn.close()

# Convert the time_group column to datetime type
df['time_group'] = pd.to_datetime(df['time_group'])

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(df['time_group'], df['average_vendor_rating'], marker='o')
plt.xlabel("Month")
plt.ylabel("Average Vendor Rating")
plt.title("Average Vendor Rating Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

12.How does the serving distance ('serving_distance') impact the average vendor rating ('vendor_rating_x')? 


```python
%%sql
SELECT
    CASE
        WHEN "serving_distance" <= 1 THEN '0 - 1 km'
        WHEN "serving_distance" <= 2 THEN '1 - 2 km'
        WHEN "serving_distance" <= 5 THEN '2 - 5 km'
        ELSE '5+ km'
    END AS distance_range,
    CAST(AVG("vendor_rating_x") AS NUMERIC(10, 2)) AS average_vendor_rating
FROM
    "Restuarant_df"
GROUP BY
    distance_range
ORDER BY
    distance_range;

```

```python

# SQL query
sql_query = """
SELECT
    CASE
        WHEN "serving_distance" <= 1 THEN '0 - 1 km'
        WHEN "serving_distance" <= 2 THEN '1 - 2 km'
        WHEN "serving_distance" <= 5 THEN '2 - 5 km'
        ELSE '5+ km'
    END AS distance_range,
    CAST(AVG("vendor_rating_x") AS NUMERIC(10, 2)) AS average_vendor_rating
FROM
    "Restuarant_df"
GROUP BY
    distance_range
ORDER BY
    distance_range;
"""
# Fetch data into a pandas DataFrame
df = pd.read_sql_query(sql_query, conn)

# Close the database connection
conn.close()

# Visualize the average vendor rating by serving distance using a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x="distance_range", y="average_vendor_rating", data=df, palette="muted")
plt.xlabel("Serving Distance")
plt.ylabel("Average Vendor Rating")
plt.title("Average Vendor Rating by Serving Distance")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
